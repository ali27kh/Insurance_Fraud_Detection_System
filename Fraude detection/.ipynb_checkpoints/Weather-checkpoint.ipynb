```python
def print_weather(data):
    ref_time = datetime.datetime.fromtimestamp(data.ref_time).strftime('%Y-%m-%d %H:%M')
    print(f"Time\t\t: {ref_time}")
    print(f"Overview\t: {data.detailed_status}")
    print(f"Wind Speed\t: {data.wind()}")
    print(f"Humidity\t: {data.humidity}")
    print(f"Temperature\t: {data.temperature('fahrenheit')}")
    print(f"Rain\t\t: {data.rain}")
    print("\n")

def get_current_weather():
    weather_api = owm.weather_at_place('Tunisia')  # Remplacez par la ville souhaitée
    weather_data = weather_api.weather  # Obtenez les données météorologiques

    print("***Current Weather***")
    print_weather(weather_data)
    print("\n")

def get_forecast_weather(date_str=None):
    print("***5 day forecast Weather***")
    for item in owm.forecast_at_place('Tunisia', '3h').forecast:
        forecast_date = datetime.datetime.fromtimestamp(item.reference_time()).strftime('%Y-%m-%d')
        if date_str and forecast_date == date_str:
            print(f"Weather on {date_str}:")
            print_weather(item)

if __name__ == '__main__':
    get_current_weather()
    
    # Si vous voulez afficher la prévision pour un jour précis (par exemple, '2025-01-27')
    get_forecast_weather('2025-01-27')  # Changez la date en fonction de vos besoins
```
