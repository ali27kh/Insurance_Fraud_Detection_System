```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

## 1-preprocessing data1

```python
data1 = pd.read_excel(r"data_client1.xlsx")
```

```python
data1
```

```python
del data1["ID"]
```

```python
# Initialize the LabelEncoder
le = LabelEncoder()

# Encode the 'fraud_reported' column
data1['exact_description_of_accidents_by_photo'] = le.fit_transform(data1['exact_description_of_accidents_by_photo'])
data1['contact_with_potential_fraudster_mechanic'] = le.fit_transform(data1['contact_with_potential_fraudster_mechanic'])
```

## 2-preprocessing data2

```python
data2 = pd.read_excel(r"data_client2.xlsx")
```

```python
data2
```

```python
data2.columns
```

```python
data2 = data2[['months_as_customer', 'age',
       'policy_number', 'policy_annual_premium',
       'capital_gains', 'capital_loss', 'incident_hour_of_the_day',
       'total_claim_amount', 'injury_claim', 'property_claim', 'vehicle_claim',
       'policy_state', 'policy_csl', 'policy_deductable', 'umbrella_limit',
       'insured_sex', 'incident_type', 'collision_type',
       'incident_severity', 'authorities_contacted', 'incident_state',
       'incident_city', 'number_of_vehicles_involved', 'property_damage',
       'bodily_injuries', 'witnesses', 'auto_year', 'police_report_available',
       'auto_make', 'auto_model', 'fraud_reported']]
```

```python
# Initialize the LabelEncoder
le = LabelEncoder()

# Encode the 'fraud_reported' column
data2['fraud_reported'] = le.fit_transform(data2['fraud_reported'])
```

```python
data2
```

```python
print(data2.columns)
```

```python
print(data2.dtypes)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = data2.select_dtypes(include=['float64', 'int64', 'int32']).columns
corr = data2[numeric_cols].corr()

fraud_corr = corr['fraud_reported'].sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Corrélation entre les variables numériques")
plt.show()

print(fraud_corr)
```

```python
#correlation > 0.012
corr_threshold = 0.012
selected_columns = fraud_corr[fraud_corr > corr_threshold].index

# Create a new DataFrame with the selected columns
data2_filtered = data2[selected_columns]
```

```python
data2_filtered
```

```python
# Concatenate data1 and data2 horizontally
data_concatenated = pd.concat([data1, data2_filtered], axis=1)
```

```python
data_concatenated.to_excel("data_client_final.xlsx", index =False)
```
