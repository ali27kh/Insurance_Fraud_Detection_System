```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import roc_auc_score
```

## 1)Clustering:

```python
data = pd.read_excel('data_client_final.xlsx')
```

```python
data.columns
```

```python
features = data.drop(columns=['fraud_reported'])
target = data['fraud_reported']

scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)

X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)

data['fraud_probability'] = model.predict_proba(features_scaled)[:, 1]

auc_score = roc_auc_score(target, data['fraud_probability'])
print(f"Score AUC : {auc_score}")
```

```python
data
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
data['fraud_cluster'] = kmeans.fit_predict(data[['fraud_probability']])

cluster_order = data.groupby('fraud_cluster')['fraud_probability'].mean().sort_values().index
cluster_mapping = {cluster: idx for idx, cluster in enumerate(cluster_order)}
data['fraud_class'] = data['fraud_cluster'].map(cluster_mapping)
```

## 2)Modeling:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import roc_curve, auc, accuracy_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
```

```python
X = data[['exact_weather_condition_by_accident',
       'exact_accident_location_per_claim',
       'number_of_findings_with_potential_fraudster',
       'exact_description_of_accidents_by_photo',
       'contact_with_potential_fraudster_mechanic', 'fraud_reported',
       'vehicle_claim', 'total_claim_amount', 'property_claim', 'injury_claim',
       'umbrella_limit', 'number_of_vehicles_involved', 'witnesses',
       'bodily_injuries', 'months_as_customer', 'policy_deductable', 'age',
       'fraud_probability']]
y = data[['fraud_class']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

param_grid_lr = {
    'C': [0.1, 1, 10],
    'solver': ['liblinear', 'saga'],
    'penalty': ['l2', 'none']
}

param_grid_knn = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

param_grid_svc = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Support Vector Machine": SVC(probability=True, random_state=42),
}
```

```python
best_models = {}

for model_name, model in models.items():
    if model_name == "Logistic Regression":
        grid_search = GridSearchCV(model, param_grid_lr, cv=5, n_jobs=-1, scoring='accuracy')
    elif model_name == "K-Nearest Neighbors":
        grid_search = GridSearchCV(model, param_grid_knn, cv=5, n_jobs=-1, scoring='accuracy')
    elif model_name == "Support Vector Machine":
        grid_search = GridSearchCV(model, param_grid_svc, cv=5, n_jobs=-1, scoring='accuracy')

    grid_search.fit(X_train, y_train)
    
    best_models[model_name] = grid_search.best_estimator_
```

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
best_model_name = max(best_models, key=lambda k: best_models[k].score(X_test, y_test))
best_model = best_models[best_model_name]
print(f"Meilleur modÃ¨le : {best_model_name} avec une prÃ©cision de {best_model.score(X_test, y_test):.4f}")

y_pred = best_model.predict(X_test)

lb = LabelBinarizer()
y_bin = lb.fit_transform(y_test)

y_prob = best_model.predict_proba(X_test)

plt.figure(figsize=(10, 8))

for i in range(y_prob.shape[1]):
    fpr, tpr, thresholds = roc_curve(y_bin[:, i], y_prob[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=2, label=f'Classe {lb.classes_[i]} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taux de faux positifs')
plt.ylabel('Taux de vrais positifs')
plt.title('Courbe ROC pour chaque classe')

plt.grid()
plt.legend(loc='lower right')


plt.show()
```

```python
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Courbe ROC (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taux de faux positifs')
plt.ylabel('Taux de vrais positifs')
plt.title('Courbe ROC')
plt.legend(loc='lower right')
plt.show()

train_sizes, train_scores, test_scores = learning_curve(best_model, X_scaled, y.values.ravel(), cv=5, n_jobs=-1)

plt.figure()
plt.plot(train_sizes, train_scores.mean(axis=1), label="Train", color='b')
plt.plot(train_sizes, test_scores.mean(axis=1), label="Test", color='r')
plt.xlabel("Taille de l'ensemble d'entraÃ®nement")
plt.ylabel("PrÃ©cision")
plt.title("Courbe d'apprentissage")
plt.legend(loc="best")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# GÃ©nÃ©ration de la matrice de confusion
y_pred = best_model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

# Affichage de la matrice de confusion sous forme de heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=lb.classes_, yticklabels=lb.classes_)
plt.xlabel('PrÃ©dictions')
plt.ylabel('VÃ©ritÃ©s observÃ©es')
plt.title('Matrice de confusion')
plt.show()
```
