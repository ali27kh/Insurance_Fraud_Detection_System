```python
import torch
from PIL import Image
```

```python
import pathlib
temp = pathlib.PosixPath
pathlib.PosixPath = pathlib.WindowsPath
```

```python
model_path = 'C:/Users/ALI/Desktop/Projets/Stages/PFE Addinn/vehicle damage detection model/code model/1st_train/weights/best.pt'
model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path, force_reload=True)

image_path = 'C:/Users/ALI/Desktop/Projets/Stages/PFE Addinn/vehicle damage detection model/code model/test5.jpg'
image = Image.open(image_path)

results = model(image)
results.show()

detections = results.pandas().xyxy[0]

for _, row in detections.iterrows():
    print(f"Class: {row['name']} | Confidence: {row['confidence']:.2f}")
```

```python
pieces_dict = {}

for _, row in detections.iterrows():
    class_name = row['name']
    
    if class_name in pieces_dict:
        pieces_dict[class_name] += 1
    else:
        pieces_dict[class_name] = 1

print("\nListe des pièces détectées avec quantités :")
for piece, quantity in pieces_dict.items():
    print(f"{piece} : {quantity}")
```

```python
import json
# Sauvegarder les résultats dans un fichier JSON
json_file_path = 'pieces_detachées.json'
with open(json_file_path, 'w', encoding='utf-8') as json_file:
    json.dump(pieces_dict, json_file, ensure_ascii=False, indent=4)

print(f"\nLes résultats ont été sauvegardés dans le fichier {json_file_path}")
```
